# Relacionando dados de Lotes Fiscais com IPTU 

Uma das formas de obter endereços da Cidade de São Paulo é relacionando a geometria do Lote Fiscal com os dados de IPTU

## Carregando geometrias de lotes

```python
import geopandas as gpd
import pandas as pd
import glob
```

```python
%load_ext autotime
```

```python
all_lotes = glob.glob("data/gis/lotes/*.zip")

li = []

for filename in all_lotes:
    zip_filename = f"zip://{filename}!{filename.replace('data/gis/lotes/', '').replace('.zip', '')}"
    print(zip_filename)
    #zipfile = "zip:///Users/name/Downloads/gadm36_AFG_shp.zip!data"
    gdf = gpd.read_file(zip_filename)
#     df = pd.read_csv(filename, encoding='latin1', 
#                      sep=';', 
#                      decimal=",", 
#                      header=None, 
#                      names=colunas[filename],
#                      skiprows=[0])
# #     li.append({"arquivo": filename,
# #                "headers": list(df.columns)})
    li.append(gdf)
```

```python
gdf_lotes = gpd.GeoDataFrame(pd.concat(li, ignore_index=True))
```

```python
gdf_lotes
```

```python
gdf_lotes.loc[gdf_lotes.drop(columns=['geometry']).duplicated(keep=False)].to_file("resultados/lotes-fiscais-duplicados.gpkg", driver='GPKG')
```

```python
gdf_lotes.drop_duplicates(subset=list(gdf_lotes.columns)[0:-1], inplace=True)
# list(gdf_lotes.columns)[0:-1]
```

```python
gdf_lotes.geometry = gdf_lotes.geometry.apply(lambda x: x.representative_point())
```

## Carregando dados do IPTU

```python
df_iptu = pd.read_csv('data/IPTU_2020.zip', sep=';',  encoding='latin1')
```

```python
colunas_endereco = ['CODLOG DO IMOVEL', 'NOME DE LOGRADOURO DO IMOVEL', 'NUMERO DO IMOVEL', 'COMPLEMENTO DO IMOVEL', 'BAIRRO DO IMOVEL']
```

```python
df_iptu.loc[:, colunas_endereco]
```

```python
df_iptu['sqlc'] = df_iptu['NUMERO DO CONTRIBUINTE'].str[0:10] + df_iptu['NUMERO DO CONDOMINIO'].str[0:2]
```

```python
df_iptu.loc[:, ['sqlc']] = df_iptu.sqlc.apply(lambda x: x if x[10:12] == '00' else x[0:6] + '0000' + x[10:12])
```

```python
df_iptu.drop_duplicates(subset=['sqlc'], inplace=True)
```
